```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
```

# Histogram Generation

```python
conf_dict = {}

frame_index = 0
with open("data.json", "r", encoding = "utf-8") as f:
  for line in f:
    try:
      obj = json.loads(line)
      for thing in obj["detections"]:
        # For every confidence value, increase count in dictionary by one
        round_value = round(thing["conf"], 2)
        conf_dict[round_value] = conf_dict.get(round_value, 0) + 1
    except Exception as e:
      print("ERROR:", e)
    print("\n")
    frame_index += 1

# Sort and display values
sorted_conf_dict = dict(sorted(conf_dict.items()))

for key, value in sorted_conf_dict.items():
  print(f"KEY: {key} | VALUE: {value}")
```

```python
# Plot Histogram
plt.xlabel("Confidence Value")
plt.ylabel("Detection Count")

plt.bar(sorted_conf_dict.keys(), sorted_conf_dict.values(), width = 0.01)
```
